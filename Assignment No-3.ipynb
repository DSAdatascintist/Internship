```python
#Q 10 Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/
import warnings
warnings.filterwarnings('ignore')
#lets first connect to web driver
driver = webdriver.Chrome(r"C:/Users/SAI BABA/Downloads/chromedriver_win32/chromedriver.exe")
driver.get('https://www.hostelworld.com/')
```

```python
search_loc=driver.find_element_by_xpath("//input[@id='search-input-field']")
search_loc.send_keys("london")
```

```python
url="https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-01-03&to=2022-01-06&guests=2&page=1"
driver.get(url)
```

```python
#extract all HOSTEL name
hostel_tags=driver.find_elements_by_xpath("//h2[@class='title title-6']")
hostel_tags[0:]
```

```python
hostel_names=[]
for i in hostel_tags:
    names=i.text
    hostel_names.append(names)
hostel_names[0:]  
```

```python
hostel_distance=[]
for i in distance_tags:
    distance=i.text
    hostel_distance.append(distance)
hostel_distance[0:]  
```

```python
#extract all ratings
rating_tags=driver.find_elements_by_xpath("//div[@class='score orange big']")
rating_tags[0:]
```

```python
rating=[]
for i in rating_tags:
    ratings=i.text
    rating.append(ratings)
rating[0:]  
```

```python
#extract total review
totalreview_tags=driver.find_elements_by_xpath("//div[@class='reviews']")
totalreview_tags[0:]
```

```python
total_review=[]
for i in totalreview_tags:
    totalreview=i.text
    total_review.append(totalreview)
total_review[0:] 
```

```python
#extract overall review
overallreview_tags=driver.find_elements_by_xpath("//div[@class='keyword']")
overallreview_tags[0:]
```

```python
overall_review=[]
for i in overallreview_tags:
    overallreview=i.text
    overall_review.append(overallreview)
overall_review[0:] 
```

```python
#extract privates from price
privates_tags=driver.find_elements_by_xpath("//div[@class='price title-5']")
privates_tags[0:]
```

```python
privates_fromprice=[]
for i in privates_tags:
    privates=i.text
    privates_fromprice.append(privates)
privates_fromprice[0:] 
```

```python
#extract dorums from price
dorums_tags=driver.find_elements_by_xpath("//div[@class='price title-5']")
dorums_tags[0:]
```

```python
dorums_fromprice=[]
for i in dorums_tags:
    dorums=i.text
    dorums_fromprice.append(dorums)
dorums_fromprice[0:] 
```

```python
#extract facility and property description
description_tags=driver.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']")
description_tags[0:]
```

```python
description=[]
for i in description_tags:
    descriptions=i.text
    description.append(descriptions)
description[0:] 
```

```python
print(len(hostel_names),len(hostel_distance),len(rating),len(total_review),len(overall_review),len(privates_fromprice),len(dorums_fromprice),len(description))
```

```python
jobs=pd.DataFrame({})
jobs['hostel']=hostel_names
jobs['distance']=hostel_distance
jobs['rating']=rating
jobs['totalreview']=total_review
jobs['overallreview']=overall_review
jobs['privates']=privates_fromprice
jobs['dorums']=dorums_fromprice
jobs['descriptions']=description
```
