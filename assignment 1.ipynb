```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
#Q1.Write a python program to display all the header tags from wikipedia.org.
#send request to the web page server
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
titles = []

for i in soup.find_all('span',class_="mw-headline"):
    titles.append(i.text)
    
titles    
```

```python
#Q2 Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
#send request to the web page server
page = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python

name = []

for i in soup.find_all('h3',class_="lister-item-header"):
    name.append(i.text)
    
name    



```

```python
year = []

for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    year.append(i.text)
    
year    
```

```python
rating = []

for i in soup.find_all('div',class_="ratings-bar"):
    rating.append(i.text)
    
rating   
```

```python
print(len(name),len(year),len(rating))
```

```python
import pandas as pd
df=pd.DataFrame({'name':name,'year':year,'rating':rating})
df
```
